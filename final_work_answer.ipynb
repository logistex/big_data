### [문제 2] 범죄 데이터프레임 정리
- [문제 1]에서 획득한 `crime_2019.txt` 파일을 읽어서
- `crime_2019` 데이터프레임에 저장하고
- 교재 1장에서 다루었던,   
'인구 규모 2018', 'CCTV 규모 2018' 자료가 저장된 csv 파일을 활용하여
- 적절한 처리 과정을 거처서 <그림 D2>와 같이 정제

|<그림 D2> crime_2019 데이터프레임의 내용|
|:--|
|![df_crime_2019](https://user-images.githubusercontent.com/10287629/102733611-0fcf0600-4381-11eb-9857-383b8d08357d.png)
|

```python
# [문제 2] 범죄 데이터프레임 정리
# 다운로드한 데이터 파일 읽어오기
crime_2019 = pd.read_csv(
    './crime_2019.txt',
     thousands=',',     # 천 단위 구분 쉼표 처리
     header = 1,        # 헤더 행 번호 지정
     sep='\t',          # 구분자 지정
)
# 자치구 합계 행을 제거 (조건으로 행 검색하여 인덱스 값을 구함)
crime_2019.drop(
    crime_2019.loc[crime_2019['자치구']=='합계'].index,
    inplace=True
)
cols = ['살인', '강도', '강간', '절도', '폭력']
# 열 이름 변경
crime_2019.rename(columns={'발생': '합계발생', '검거': '합계검거'}, inplace=True)
for i, c in enumerate(cols):
    crime_2019.rename(columns={'발생.'+str(i+1): c, '검거.'+str(i+1): c+'검거'}, inplace=True)
# 기존 합계 및 기간 열 제거
crime_2019.drop(columns=['합계발생', '합계검거', '기간'], inplace=True)
# 검거율 계산하고, 검거 열 제거
for c in cols:
    crime_2019[c+'검거율'] = crime_2019[c+'검거'] / crime_2019[c] * 100
    crime_2019.drop(columns=[c+'검거'], inplace=True)
# 자치구 열을 인덱스로 지정
crime_2019.set_index('자치구', inplace=True)

# 교재1장에서 작성했던 2018년도 기준 인구수와 CCTV 소계 자료 읽어오기
pop_cctv_2018 = pd.read_csv('./CCTV_result_new.csv',
                          encoding='UTF-8',
                          index_col='자치구')
# 범죄 데이터에 1장의 인구 및 CCTV 데이터 추가
crime_2019[['인구 규모', 'CCTV 규모']] = pop_cctv_2018[['인구 규모 2018', 'CCTV 규모 2018']]
crime_2019
```

### [문제 3] 범죄 발생 건수 정규화
- `crime_2019`  데이터프레임에서 범죄 발생 수치를
- 자치구 간의 상대적 수치인 (0, 1) 범위로 정규화하여,
- `crime_2019_norm` 데이터프레임으로 저장

|<그림 D3> 범죄 발생 건수를 자치구별 상대적 수치로 정규화한 결과|
|:--|
|![df_crime_2019_norm](https://user-images.githubusercontent.com/10287629/102707122-9cbf8400-42db-11eb-84f8-8cbc6e89d440.png)|

```python
# [문제 3] 범죄 발생 건수 정규화
from sklearn import preprocessing

crime_2019_norm = pd.DataFrame(
    index=crime_2019.index,
    columns=cols,
    data=preprocessing.MinMaxScaler().fit_transform(
        crime_2019[cols].values.astype(float)).round(2)
)
crime_2019_norm
```

### [문제 4] 범죄 발생 비율 계산
- `crime_2019` 데이터프레임의
- 범죄 건수에 인구 자료를 적용하여 
- `crime_2019_per_capita`에 <그림 D4>와 같이
- 인구 십만 명 당 범죄 발생 비율을 저장

|<그림 D4> 범죄 발생 건수를 자치구별 (인구 십만 명 당) 비율로 계산한 결과|
|:--|
|![df_crime_2019_per_capita](https://user-images.githubusercontent.com/10287629/102707124-9cbf8400-42db-11eb-95bf-1196f875e059.png)|

```python
# [문제 4] 범죄 발생 비율 계산
# crime_2019에서 범죄 건수만 crime_2019_per_capita로 복사
crime_2019_per_capita = pd.DataFrame(
    index=crime_2019.index,
    columns=cols,
    data=crime_2019[cols]
)
# 인구 대비 범죄 현황 계산
for c in cols:
    crime_2019_per_capita[c] = (
        crime_2019_per_capita[c] / crime_2019['인구 규모'] * 100000
    ).round(2)  # 십만명 당 범죄 건수
crime_2019_per_capita
```

```python
# 시각화 준비 작업
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)  # 경고 메시지 출력 억제
import folium
from folium.features import DivIcon

# 지도 json 파일 읽어오기
import json
geo_path = './skorea_municipalities_geo_simple.json'
geo_str = json.load(open(geo_path, encoding='utf-8'))

# 함수 정의
def crime_map(crime, rate=False, mark_circle=False):
    crime_english = {'살인':'Murder', '강도':'Robbery', '강간':'Rape', '폭력':'Violence', '절도':'Theft'}
    map = folium.Map(
        location=[37.5502, 126.982],
        zoom_start=11,
        )
    if not rate:  # 건수 시각화인 경우
        my_data = crime_2019[crime]
        my_cols = [crime_2019_norm.index, crime_2019_norm[crime]]
        my_text = crime_2019_norm.index
        title_html_1 = f'<h3 align="center" style="font-size:15px"><b><u>Cases of {crime_english[crime]}</u></b></h3>'
        title_html_2 = '<h5 align="center" style="font-size:13px"><b>Number of cases</b>(Standadized Number)</h5>'   
    else:         # 비율 시각화인 경우
        my_data = crime_2019_per_capita[crime]  # 십만명 당 범죄 비율
        my_cols = [crime_2019_per_capita.index, my_data]
        my_text = crime_2019_per_capita.index
        title_html_1 = f'<h3 align="center" style="font-size:15px"><b><u>Rates of {crime_english[crime]}</u></b></h3>'
        title_html_2 = '<h5 align="center" style="font-size:13px"><b>Rate of cases per 100,000 population</b>(Standadized Number)</h5>'        
    if mark_circle:  # 과제 정의서에 누락된 내용이지만 추가하였음
        title_html_1 = title_html_1.replace('</u>', ' with Circle Mark Proportional to Arrest Power</u>')

    map.choropleth(geo_data = geo_str,
                   data = my_data,
                   columns = my_cols,  # 건수 기준이면 정규화된 건수로, 비율 기준이면 그냥 비율로 적용
                   fill_color = 'PuRd',    # PuRd, YlGnBu
                   key_on = 'feature.id')
    map.get_root().html.add_child(folium.Element(title_html_1 + title_html_2))  # 차트에 제목 추가

    for n, ndx in enumerate(crime_2019_norm.index):  # 자치구 정보 출력
        if rate:  # 건수는 정수로, 비율은 소수 두자리까지 출력
            my_html = f'<div style="font-size: 9pt">{my_text[n]}<br>' \
                + f'<b>{my_data[n]:.2f}</b>({crime_2019_norm[crime][n]:.2f})</div>'
        else:
            my_html=f'<div style="font-size: 9pt">{my_text[n]}<br>' \
                + f'<b>{my_data[n]}</b>({crime_2019_norm[crime][n]:.2f})</div>'
        text = my_text[n]
        folium.map.Marker(
            [office['위도'][n], office['경도'][n]],       
            icon=DivIcon(
                icon_size=(150, 40),
                icon_anchor=(20, 20),
                html=my_html,
            )
        ).add_to(map)

    if mark_circle:  # 건수(또는 비율)과 정규화 수치를 출력
        r_name = crime + '검거율'
        for n in office.index:
            folium.CircleMarker(
                [office.위도[n], office.경도[n]],
                radius = np.sqrt(crime_2019[r_name][n]) * 3.0,  # 원 면적에 비례하도록 np.sqrt() 적용
                color='#3186cc', fill_color='#3186cc', fill=True,
                popup=f'arrest rate: {crime_2019[r_name][n]:.2f}',  # popup 수치에도 소숫점 2자리 적용 
                ).add_to(map)
            
    return map
```

      <u>꺽은선 레이블과 하이차트 메뉴 아이콘이 나타나게 하려면  
      어떻게 해야 하는지</u>에 관한 고민이 필요했음
    - 이 점을 해결한 일부 학생들의 성과에 박수를!
    - 일부 학생들은  
      <그림 V2>에서는 이 문제를 해결했으나,  
      <그림 V1>에서는 이 문제를 굳이 외면했는데 ...  
      과제 정의서에 충실하려는 마음이었으리라 추측하지만,   
      이는 너무 경직된 접근 방식이 아닐까요?
- 해결 방법
    - 하이차트에서 꺽은선 레이블을 표시하는 방법은  
      `series-label.js` 라이브러리를 포함시키는 것임
    - 하이차트에서 메뉴 아이콘을 표시하는 방법은  
      `exporting.js` 라이브러리를 포함시키는 것임
    - 이 두 방법은 템플릿 파일에 하이차트 라이브러리를 코드에 포함시키는 것과 동일한 방식으로 처리함

    - 하이차트의 기능은 모듈화가 잘 되어 있고,  
      이런 방식으로 해당 기능을 담당하는 라이브러리를 포함시키는 방법으로 구현이 가능함
