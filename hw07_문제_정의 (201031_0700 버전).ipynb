### 두 연도를 단일 차트로 시각화
- <그림 12>는 2015년도 산점도와 2018년도 산점도를 단일 차트로 통합하여 시각화한 결과이다.
- 2018년도 원은 투명도를 0%로 적용하여 강조하였으나,
  2015년도 원은 투명도를 50%로 적용하여 순화하였다.
- 그리고 2015년도 원의 중심에서 2018년도 원의 중심으로 붉은 선을 그렸다.  
  붉은 선의 강도 및 두께는 원의 중심 간 거리에 비례하도록 처리하였다.  
  선명하고 두꺼울수록 2015년도에 비하여 크게 변화한 것이다.
- 결과적으로 희미한 2015년도의 원에서 선을 따라 올라가면 선명한 2018년도의 원을 쉽게 추적할 수 있다.
- 자치구의 이동 궤적이 오른쪽으로 상승하면 해당 자치구의 인구가 증가한 경우이고,  
  자치구의 이동 궤적이 왼쪽으로 상승하면 해당 자치구의 인구가 감소한 경우이다.
- 2015년도 추세선은 연한 점선으로, 2018년도 추세선은 진한 실선으로 처리하였다.
- <그림 12>를 시각화하는 부분이 문제 7에 해당하며, 세부적인 채점 기준은 다음과 같다.
    - (1.0 점) 2018년 데이터에 해당하는 거품(원) 시각화의 정확도
    - (1.0 점) 2015년 데이터에 해당하는 거품(원) 시각화의 정확도
    - (2.0 점) 2015년 거품과 2018년 거품을 연결하는 선 시각화의 정확도
- <그림 12>를 시각화할 때 필요한 핵심 사항은 다음과 같다.

    - 2018년도 데이터로 산점도, 추세선 및 마커 레이블을 시각화하되, 투명 효과를 배제
    - 2015년도 거품에서 2018년도 거품으로 이동 궤적을 표시하는 선을 자치구별로 반복하여 시각화
